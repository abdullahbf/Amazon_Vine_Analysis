```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Import vine_table.csv
df = pd.read_csv("Resources/vine_table.csv")
df.head()
```

```python
# DataFrame containing rows where total_votes >= 20
votes_twentyplus_df = df.loc[df["total_votes"] >= 20]
votes_twentyplus_df.head()
```

```python
# DataFrame containing all the rows where (helpful_votes/total_votes) >= 0.50
fiftyorgreater_df = votes_twentyplus_df.loc[votes_twentyplus_df["helpful_votes"]/votes_twentyplus_df["total_votes"] >= 0.50]
fiftyorgreater_df.head()
```

```python
# Retrieve rows where vine == 'Y' from fiftyorgreater_df
vine_df = fiftyorgreater_df.loc[fiftyorgreater_df["vine"] == "Y"]
vine_df.head()
```

```python
# Retrieve rows where vine == 'N' from fiftyorgreater_df
not_vine_df = fiftyorgreater_df.loc[fiftyorgreater_df["vine"] == "N"]
not_vine_df.head()
```

```python
# Function to caluculate total number of reviews, number of five star reviews and percentage of five star reviews 
# for vine_df (paid) and not_vine_df (unpaid)
def review_details(DataFrame):
    
    total_reviews = DataFrame["review_id"].count()
    five_stars_df = DataFrame[DataFrame["star_rating"] == 5.0]
    five_stars_reviews = five_stars_df["review_id"].count()
    five_stars_percentage = (five_stars_reviews/total_reviews) * 100
    
    print(f"""
    Total number of reviews = {total_reviews}
    Number of five star reviews = {five_stars_reviews} 
    Percentage of five star reviews = {five_stars_percentage:,.2f}%
    """)
```

```python
review_details(vine_df)
```

```python
review_details(not_vine_df)
```
